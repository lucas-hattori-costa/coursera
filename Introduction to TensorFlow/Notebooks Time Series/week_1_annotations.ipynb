# Time Series with Tensorflow

## Caracter√≠sticas de uma time series

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
```

```python
def plot_series(time, series, format="-", start=0, end=None, label=None):
    plt.plot(time[start:end], series[start:end], format, label=label)
    plt.xlabel("Time")
    plt.ylabel("Value")
    if label:
        plt.legend(fontsize=14)
    plt.grid(True)
```

Defining a trend for a ts

```python
def trend(time, slope = 0):
    return slope * time
```

```python
# Simple ts with positive trend
time = np.arange(4 * 365 + 1)
baseline = 10
series = trend(time, 0.1)

plt.figure(figsize=(10, 6))
plot_series(time, series)
plt.show()
```

Defining seasonal patterns for our ts

```python
def seasonal_pattern(season_time):
    """Just an arbitrary pattern, you can change it if you wish"""
    return np.where(season_time < 0.4,
                    np.cos(season_time * np.pi),
                    1 / np.exp(3 * season_time))

def seasonality(time, period, amplitude=1, phase=0):
    """Repeats the same pattern at each period"""
    season_time = ((time + phase) % period) / period
    return amplitude * seasonal_pattern(season_time)
```

```python
def second_seasonal_pattern(season_time):
    '''Another arbitrary pattern'''
    return np.where(season_time < 0.7,
                    np.exp(season_time * 0.33),
                    np.cos(season_time * 0.5 * np.pi))

def combine_seasonality(time, period, amplitude = 1, phase = 0):
    '''Combines both seasonal patterns'''
    season_time = ((time + phase) % period) / period
    return amplitude * seasonal_pattern(season_time) * second_seasonal_pattern(season_time)
```

```python
baseline = 10
amplitude = 40
series = combine_seasonality(time, period=365, amplitude=amplitude)

plt.figure(figsize=(10, 6))
plot_series(time, series)
plt.show()
```

Combining trend and seasonality

```python
slope = 0.05
series = baseline + trend(time, slope) + combine_seasonality(time, period=365, amplitude=amplitude)

plt.figure(figsize=(10, 6))
plot_series(time, series)
plt.show()
```

Creating white noise

```python
def white_noise(time, noise_level=1, seed=None):
    rnd = np.random.RandomState(seed)
    return rnd.randn(len(time)) * noise_level
```

```python
noise = white_noise(time, noise_level = 4, seed=42)

plt.figure(figsize=(10, 6))
plot_series(time, noise)
plt.show()
```

```python
# adding the noise to the ts
series += noise

plt.figure(figsize=(10, 6))
plot_series(time, series)
plt.show()
```
